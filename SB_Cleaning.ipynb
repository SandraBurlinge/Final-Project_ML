# Data Cleaning

```python
birthRate_df = pd.read_csv('Birth_Rate.csv')
birthRate_df.head()
```

```python
Population_df = pd.read_csv('Population.csv')
Population_df.head()
```

```python
Households_df = pd.read_csv('Households.csv')
Households_df.head()
```

```python
# Merge 2 out of 3 DataFrames
df1 = birthRate_df.merge(Population_df, on=['Years','States'], how='left')
df1.head(5)
```

```python
len(list(df1))
```

```python
# Merge 3 out of 3 DataFrames
df2 = df1.merge(Households_df, on=['Years','States'], how='left')
df2.head(5)
```

```python
df2.to_csv('Resources/cleaned_data2.csv',
                  encoding="utf-8", index=False, header=True)
```

```python
df3=df2.drop(columns=['Families percent below poverty level',
             'General fertility rate',
             'Total fertility rate'])
```
