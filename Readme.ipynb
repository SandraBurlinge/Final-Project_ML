## Project Description:

The project aims to xplore the relationship between Birth Rate and wider macroeconomicindicators across time (from 2007 to 2017) and geography (state). As well as determine the most important feature set that would enable the prediction of the birth rate across the states such as:

* Total households & median incomes
* Marriage ratio
* Educational attainment
* Women in labor force
* Teenage pregnancy
* Total population
* Total sex ratio
* Total households & Households with own children

## Predictive modeling using Machine Learning

To train the model and make the prediction of birth rate I chose Multiple Linear Regression as for my data worked best. To train and test the model I had followed the regular steps as follows:
* Preprocess the dataset prior to fitting the model.
* Perform feature selection and remove unnecessary features
* Started by cleaning the data, removing unnecessary columns if needed, and scaling the data.

Not all variables were significant, some as 'Total women with births' or 'Families percent below poverty level' had to be removed as they were insignificant.  After trying different variations fitting different variables I end up keeping 14 features (that includes education attainment that splits into additional 4 subcategories (Less than high school graduate, college or associate's degree, Bachelor's degree, Graduate degree).


## Regression Results

The test score I got is pretty low. There was not much variation when these measures were calculated in training model either. The highest coefficient I had got was Marriage ratio: 0.654 and the second best Total sex ratio: 0.460. 

To evaluate the overall fit of a linear model, I use theR-squaredvalue and Mean Squared Error.

The model roughly accounts for 68% of the variability of the data. Which overall means this model is not good and accurate to make the predictions.
It also can mean that this model might be missing significant independent variables or needs to be replaced with a higher coefficient score variables.
